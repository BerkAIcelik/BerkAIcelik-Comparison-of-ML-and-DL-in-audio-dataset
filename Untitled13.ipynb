```python
import os
import librosa
import numpy as np
import pandas as pd
from scipy.signal import stft
import scipy.stats

# Ana klas√∂r yolu
base_path = "/content/drive/MyDrive/KAYIT_DOSYALARI"
ignore_folder = "inohom"  # Kullanƒ±lmayacak klas√∂r

# STFT parametreleri
fs = 16000  # √ñrnekleme frekansƒ±
nperseg = 2048
noverlap = int(nperseg * 0.75)  # %75 √∂rt√º≈üme

print(f"nperseg: {nperseg}, noverlap: {noverlap}")

# Frekans bantlarƒ± (100 Hz geni≈ülik, 50 Hz √∂rt√º≈üme, 0-8000 Hz arasƒ±)
bands = {chr(97+i): (i*50, i*50 + 100) for i in range(16)}

# STFT √∂zelliklerini tutacak listeler
zaman_features = []
frekans_features = []
zaman_frekans_features = []

# Klas√∂rlerin listesini al
folders = [folder for folder in os.listdir(base_path) if folder != ignore_folder and os.path.isdir(os.path.join(base_path, folder))]

# Her klas√∂r ve dosya i√ßin d√∂ng√º
for folder in folders:
    folder_path = os.path.join(base_path, folder)
    files = [file for file in os.listdir(folder_path) if file.endswith(".wav")]

    for file in files:
        file_path = os.path.join(folder_path, file)

        try:
            # Ses dosyasƒ±nƒ± y√ºkle
            y, sr = librosa.load(file_path, sr=None)

            # üìå **STFT Hesaplama**
            f, t, Zxx = stft(y, fs=sr, nperseg=nperseg, noverlap=noverlap)
            magnitude_spectrum = np.abs(Zxx)

            # üìå **Zaman Domeni √ñznitelikleri**
            rms_energy = np.mean(librosa.feature.rms(y=y))
            zero_crossing_rate = np.mean(librosa.feature.zero_crossing_rate(y=y))
            mean_amplitude = np.mean(y)
            max_amplitude = np.max(y)
            min_amplitude = np.min(y)
            median_amplitude = np.median(y)
            std_amplitude = np.std(y)
            silence_threshold = 0.01  # Sessizlik e≈üiƒüi
            silence_ratio = np.sum(np.abs(y) < silence_threshold) / len(y)
            crest_factor = max_amplitude / (rms_energy + 1e-12)#tepe fakt√∂r√º
            autocorr = np.correlate(y, y, mode='full')
            autocorr_max = np.max(autocorr)

            zaman_dict = {
                "Folder": folder,
                "File": file,
                "RMS_Energy": rms_energy,
                "Zero_Crossing_Rate": zero_crossing_rate,
                "Mean_Amplitude": mean_amplitude,
                "Max_Amplitude": max_amplitude,
                "Min_Amplitude": min_amplitude,
                "Median_Amplitude": median_amplitude,
                "Std_Amplitude": std_amplitude,
                "Silence_Ratio": silence_ratio,
                "Crest_Factor": crest_factor,
                "autocorr_max": autocorr_max
            }
            zaman_features.append(zaman_dict)

            # üìå **Frekans Domeni √ñznitelikleri**
            spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr))
            spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr))
            spectral_flatness = np.mean(librosa.feature.spectral_flatness(y=y))
            spectral_contrast = np.mean(librosa.feature.spectral_contrast(S=magnitude_spectrum, sr=sr))
            mfcc = np.mean(librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13), axis=1)
                # **Yeni Spektral √ñznitelikler**
            stft_magnitude = np.abs(librosa.stft(y))
            spectral_skewness = scipy.stats.skew(stft_magnitude.flatten())  # Eƒüiklik
            spectral_kurtosis = scipy.stats.kurtosis(stft_magnitude.flatten())  # Basƒ±klƒ±k
            spectral_rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr))  # Roll-off

            frekans_dict = {
                "Folder": folder,
                "File": file,
                "Spectral_Centroid": spectral_centroid,
                "Spectral_Bandwidth": spectral_bandwidth,
                "Spectral_Flatness": spectral_flatness,
                "Spectral_Contrast": spectral_contrast,
                "Spectral_Skewness": spectral_skewness,
                "Spectral_Kurtosis": spectral_kurtosis,
                "Spectral_Rolloff": spectral_rolloff,
                "Peak_Frequency": peak_frequency,

            }
            for i, val in enumerate(mfcc):
                frekans_dict[f"MFCC_{i+1}"] = val

            frekans_features.append(frekans_dict)

            # üìå **Zaman-Frekans √ñznitelikleri (STFT Geni≈ületildi)**
            stft_dict = {
                "Folder": folder,
                "File": file,
                "STFT_Mean_Mag": np.mean(magnitude_spectrum),
                "STFT_Var_Mag": np.var(magnitude_spectrum),
                "STFT_Energy_Mag": np.sum(magnitude_spectrum**2),
                "STFT_Median_Mag": np.median(magnitude_spectrum),
                "STFT_Max_Mag": np.max(magnitude_spectrum),
                "STFT_Min_Mag": np.min(magnitude_spectrum),
                "STFT_Skew_Mag": pd.Series(magnitude_spectrum.flatten()).skew(),
                "STFT_Kurt_Mag": pd.Series(magnitude_spectrum.flatten()).kurtosis(),
                "STFT_Entropy_Mag": -np.sum(magnitude_spectrum * np.log(magnitude_spectrum + 1e-12)),
                "STFT_Var_Time": np.var(magnitude_spectrum, axis=1).mean(),
                "STFT_Var_Freq": np.var(magnitude_spectrum, axis=0).mean(),
            }

            # üìå **Frekans Bantlarƒ±na G√∂re STFT √ñznitelikleri**
            for band_name, band_range in bands.items():
                band_indices = np.where((f >= band_range[0]) & (f < band_range[1]))[0]
                if len(band_indices) > 0:
                    band_power = np.sum(magnitude_spectrum[band_indices, :])
                    max_power = np.max(magnitude_spectrum[band_indices, :])
                    min_power = np.min(magnitude_spectrum[band_indices, :])
                    std_power = np.std(magnitude_spectrum[band_indices, :])
                    median_power = np.median(magnitude_spectrum[band_indices, :])
                    stft_var_time = np.var(magnitude_spectrum, axis=1).mean()  # Zaman boyunca varyans
                    stft_var_freq = np.var(magnitude_spectrum, axis=0).mean()  # Frekans boyunca varyans
                    peak_frequency = f[band_indices[np.argmax(magnitude_spectrum[band_indices, :].sum(axis=1))]] if len(band_indices) > 0 else 0
                    mean_frequency = np.sum(f[band_indices] * magnitude_spectrum[band_indices, :].sum(axis=1)) / np.sum(magnitude_spectrum[band_indices, :])

                    stft_dict[f"{band_name}_Band_Power"] = band_power
                    stft_dict[f"{band_name}_Max_Power"] = max_power
                    stft_dict[f"{band_name}_Min_Power"] = min_power
                    stft_dict[f"{band_name}_Std_Power"] = std_power
                    stft_dict[f"{band_name}_Median_Power"] = median_power
                    stft_dict[f"{band_name}_Peak_Frequency"] = peak_frequency
                    stft_dict[f"{band_name}_Mean_Frequency"] = mean_frequency
                    stft_dict[f"{band_name}_Var_Time"] = stft_var_time
                    stft_dict[f"{band_name}_Var_Frequency"] = stft_var_freq

            zaman_frekans_features.append(stft_dict)

        except Exception as e:
            print(f"Hata olu≈ütu: {file_path}, {e}")

# üìå **√ñznitelikleri CSV olarak kaydet**
zaman_df = pd.DataFrame(zaman_features)
frekans_df = pd.DataFrame(frekans_features)
zaman_frekans_df = pd.DataFrame(zaman_frekans_features)

zaman_df.to_csv("zaman_all_audio_features.csv", index=False)
frekans_df.to_csv("frekans_all_audio_features.csv", index=False)
zaman_frekans_df.to_csv("ikisi_all_audio_features.csv", index=False)

# üìå **T√ºm √∂znitelikleri birle≈ütir ve kaydet**
all_features_df = pd.merge(zaman_df, frekans_df, on=["Folder", "File"], how="inner")
all_features_df = pd.merge(all_features_df, zaman_frekans_df, on=["Folder", "File"], how="inner")
all_features_df.to_csv("all_audio_features.csv", index=False)

print("\n‚úÖ STFT ile geni≈ületilmi≈ü √∂znitelikler ba≈üarƒ±yla kaydedildi!")
```

```python
# üìå Google Colab i√ßin GPU Kullanƒ±mƒ±nƒ± Etkinle≈ütir
import tensorflow as tf
import os

# GPU kullanƒ±labilir mi kontrol et
print("TensorFlow GPU Kullanƒ±labilirliƒüi:", tf.config.list_physical_devices('GPU'))

# GPU belleƒüini dinamik olarak ayarla (Crash √∂nleme)
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("‚úÖ GPU belleƒüi dinamik olarak ayarlandƒ±.")
    except RuntimeError as e:
        print("‚ùå GPU bellek ayarƒ± yapƒ±lamadƒ±:", e)

# üìå TensorFlow'un ilk GPU'yu kullanmasƒ±nƒ± zorunlu kƒ±l
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

# üìå Gerekli k√ºt√ºphaneler
import numpy as np
import pandas as pd
import time
import pickle
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, LSTM, Dropout, MaxPooling2D, BatchNormalization, Reshape
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# üìå 1. **Veriyi Y√ºkle ve Stratified Train-Test Split**
def load_and_preprocess_data(input_path="/content/drive/MyDrive/featuresforsignal/all_audio_features.csv"):
    data = pd.read_csv(input_path)

    # Giri≈ü (X) ve √ßƒ±kƒ±≈ü (y) ayrƒ±mƒ±
    X = data.drop(columns=["Folder", "File"]).values  # √ñzellikler
    y = data["Folder"].values  # Sƒ±nƒ±f Etiketleri

    # **Etiketleri Sayƒ±sal Hale Getir (Label Encoding)**
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    # Veriyi stratified olarak eƒüitim ve test setlerine ayƒ±r
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

    # **2D CNN i√ßin veriyi uygun ≈üekle getir**
    X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1, 1)  # (√∂rnek sayƒ±sƒ±, √∂zellik sayƒ±sƒ±, kanal, derinlik)
    X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1, 1)

    # **LSTM i√ßin veriyi uygun ≈üekle getir**
    X_train_lstm = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)  # (√∂rnek sayƒ±sƒ±, zaman adƒ±mƒ±, √∂zellik sayƒ±sƒ±)
    X_test_lstm = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    return X_train_cnn, X_test_cnn, X_train_lstm, X_test_lstm, y_train, y_test, label_encoder

X_train_cnn, X_test_cnn, X_train_lstm, X_test_lstm, y_train, y_test, label_encoder = load_and_preprocess_data()

# üìå 2. **2D CNN Modelini Tanƒ±mla**
from tensorflow.keras.optimizers import Adam

def build_cnn_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(64, (5, 5), activation='relu', padding="same", input_shape=input_shape),
        BatchNormalization(),
        MaxPooling2D((2, 1)),  # **(2, 2) yerine (2, 1) kullandƒ±k**

        Conv2D(128, (5, 5), activation='relu', padding="same"),
        BatchNormalization(),
        MaxPooling2D((2, 1)),

        Conv2D(256, (3, 3), activation='relu', padding="same"),
        BatchNormalization(),
        MaxPooling2D((2, 1)),

        Flatten(),
        Dense(512, activation='relu'),
        Dropout(0.3),
        Dense(256, activation='relu'),
        Dropout(0.2),
        Dense(num_classes, activation='softmax')
    ])

    optimizer = Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model




# üìå 3. **LSTM Modelini Tanƒ±mla**
from tensorflow.keras.layers import LSTM, Bidirectional

def build_lstm_model(input_shape, num_classes):
    model = Sequential([
        Bidirectional(LSTM(256, return_sequences=True, activation='tanh', dropout=0.3), input_shape=input_shape),
        Bidirectional(LSTM(128, return_sequences=True, activation='tanh', dropout=0.3)),
        LSTM(64, activation='tanh', dropout=0.1),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(28, activation='relu'),
        Dropout(0.2),
        Dense(num_classes, activation='softmax')
    ])

    optimizer = Adam(learning_rate=0.00001)  # **√ñƒürenme Oranƒ± G√ºncellendi**
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model


# üìå 4. **GPU Kullanarak CNN Modelini Eƒüit ve Kaydet**
with tf.device('/GPU:0'):
    cnn_model = build_cnn_model(input_shape=(X_train_cnn.shape[1], 1, 1), num_classes=len(label_encoder.classes_))
    cnn_model.fit(X_train_cnn, y_train, epochs=120, batch_size=64, validation_data=(X_test_cnn, y_test))
    cnn_model.save("cnn_model.h5")
    print("\n‚úÖ CNN Modeli Kaydedildi: cnn_model.h5")
# üìå 5. **GPU Kullanarak LSTM Modelini Eƒüit ve Kaydet**
with tf.device('/GPU:0'):
    lstm_model = build_lstm_model(input_shape=(X_train_lstm.shape[1], 1), num_classes=len(label_encoder.classes_))
    lstm_model.fit(X_train_lstm, y_train, epochs=100, batch_size=64, validation_data=(X_test_lstm, y_test))
    lstm_model.save("lstm_model.h5")
    print("\n‚úÖ LSTM Modeli Kaydedildi: lstm_model.h5")




# üìå 6. **Test Setinde Modelleri √áalƒ±≈ütƒ±r ve S√ºreyi √ñl√ß**
cnn_start_time = time.time()
cnn_predictions = np.argmax(cnn_model.predict(X_test_cnn), axis=1)
cnn_end_time = time.time()

lstm_start_time = time.time()
lstm_predictions = np.argmax(lstm_model.predict(X_test_lstm), axis=1)
lstm_end_time = time.time()

cnn_test_time = cnn_end_time - cnn_start_time
lstm_test_time = lstm_end_time - lstm_start_time

# üìå 7. **Test S√ºrelerini Kaydet**
with open("test_times.txt", "w") as f:
    f.write(f"CNN Test Time: {cnn_test_time:.4f} seconds\n")
    f.write(f"LSTM Test Time: {lstm_test_time:.4f} seconds\n")

print(f"\n‚è≥ CNN Test Time: {cnn_test_time:.4f} seconds")
print(f"‚è≥ LSTM Test Time: {lstm_test_time:.4f} seconds")


# üìå 8. **Confusion Matrix √áiz**
def plot_confusion_matrix(y_true, y_pred, title):
    conf_matrix = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12, 8))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Tahmin Edilen")
    plt.ylabel("Ger√ßek")
    plt.show()

plot_confusion_matrix(y_test, cnn_predictions, "CNN")
plot_confusion_matrix(y_test, lstm_predictions, "LSTM")
```

```python
import os
import pandas as pd
import numpy as np
from scipy.signal import stft

# üìå Veri klas√∂r√º
data_dir = "/content/drive/MyDrive/output"

# üìå STFT Parametreleri (Senin Dosyalarƒ±na G√∂re G√ºncellendi)
fs = 90  # √ñrnekleme frekansƒ±
nperseg = int(2 * fs)  # √ñrnekleme frekansƒ±nƒ±n 2 katƒ±
noverlap = int(nperseg * 0.75)  # %75 √∂rt√º≈üme

# üìå Saƒülƒ±k durumu belirleme fonksiyonlarƒ±
def extract_health_status(filename):
    return "Healthy" if "healthy" in filename.lower() else "Unhealthy"

def extract_damage_type(filename):
    if "healthy" in filename.lower():
        return "None"
    else:
        damage_types = ["notch", "scracth", "singlecut", "twocut", "warped"]
        for damage in damage_types:
            if damage in filename.lower():
                return damage.capitalize()
        return "Unknown"

# üìå T√ºm CSV dosyalarƒ±nƒ± al
csv_files = [f for f in os.listdir(data_dir) if f.endswith(".csv")]

# üìå √ñznitelikleri saklamak i√ßin liste
all_features = []

# üìå Her dosya i√ßin i≈ülem yap
for file in csv_files:
    file_path = os.path.join(data_dir, file)

    try:
        # **CSV dosyasƒ±nƒ± oku**
        df = pd.read_csv(file_path)

        # **Saƒülƒ±k durumu ve hasar t√ºr√ºn√º belirle**
        health_status = extract_health_status(file)
        damage_type = extract_damage_type(file)

        # **ƒ∞≈ülenecek sens√∂r verileri**
        sensor_columns = ['Voice', 'Acceleration X', 'Acceleration Y', 'Acceleration Z',
                          'Gyro X', 'Gyro Y', 'Gyro Z', 'Temperature']

        # üìå **√ñznitelik √ßƒ±karma**
        features = {"File Name": file, "Health Status": health_status, "Damage Type": damage_type}

        for col in sensor_columns:
            if col in df.columns:
                # **STFT Hesaplama**
                f, t, Zxx = stft(df[col].values, fs=fs, nperseg=nperseg, noverlap=noverlap)
                magnitude_spectrum = np.abs(Zxx)

                # üìå **STFT Spektral √ñzellikleri**
                features[f'{col}_mean_mag'] = np.mean(magnitude_spectrum)
                features[f'{col}_var_mag'] = np.var(magnitude_spectrum)
                features[f'{col}_energy_mag'] = np.sum(magnitude_spectrum**2)
                features[f'{col}_median_mag'] = np.median(magnitude_spectrum)
                features[f'{col}_max_mag'] = np.max(magnitude_spectrum)
                features[f'{col}_min_mag'] = np.min(magnitude_spectrum)
                features[f'{col}_skew_mag'] = pd.Series(magnitude_spectrum.flatten()).skew()
                features[f'{col}_kurt_mag'] = pd.Series(magnitude_spectrum.flatten()).kurtosis()
                features[f'{col}_entropy_mag'] = -np.sum(magnitude_spectrum * np.log(magnitude_spectrum + 1e-12))

        # üìå **√ñzellikleri listeye ekleyelim**
        all_features.append(features)

    except Exception as e:
        print(f"Hata olu≈ütu: {file_path}, {e}")

# üìå **T√ºm √∂zellikleri bir DataFrame olarak kaydet**
features_df = pd.DataFrame(all_features)
features_df.to_csv("dron_all_audio_features.csv", index=False)

print("\n‚úÖ √ñznitelik √ßƒ±karma i≈ülemi ba≈üarƒ±yla tamamlandƒ±!")
```

```python
import os
import numpy as np
import pandas as pd
import time
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, LSTM, Dropout, MaxPooling2D, BatchNormalization, Bidirectional
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# üìå 1. **GPU Kullanƒ±mƒ±nƒ± Etkinle≈ütir**
print("TensorFlow GPU Kullanƒ±labilirliƒüi:", tf.config.list_physical_devices('GPU'))
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("‚úÖ GPU belleƒüi dinamik olarak ayarlandƒ±.")
    except RuntimeError as e:
        print("‚ùå GPU bellek ayarƒ± yapƒ±lamadƒ±:", e)

os.environ["CUDA_VISIBLE_DEVICES"] = "0"

# üìå 2. **Veriyi Y√ºkle ve ƒ∞≈üle**
def load_and_preprocess_data(input_path="dron_all_audio_features.csv"):
    data = pd.read_csv(input_path)

    # üìå **Eksik "Health Status" ve "Damage Type" deƒüerlerini d√ºzelt**
    data["Health Status"].fillna("Healthy", inplace=True)
    data["Damage Type"].fillna("None", inplace=True)

    # **X ve Y ayrƒ±mƒ± (Saƒülƒ±klƒ±/Saƒülƒ±ksƒ±z sƒ±nƒ±flandƒ±rmasƒ± i√ßin)**
    X = data.drop(columns=["File Name", "Health Status", "Damage Type"]).values  # √ñzellikler
    y = data["Health Status"].values  # √áƒ±kƒ±≈ü (Saƒülƒ±klƒ± / Saƒülƒ±ksƒ±z)

    # üìå **Etiketleri Sayƒ±sal Hale Getir (Label Encoding)**
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    # **Veriyi stratified olarak eƒüitim ve test setlerine ayƒ±r**
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

    # üìå **CNN i√ßin veriyi uygun hale getir**
    X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1, 1)
    X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1, 1)

    # üìå **LSTM i√ßin veriyi uygun hale getir**
    X_train_lstm = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test_lstm = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    return X_train_cnn, X_test_cnn, X_train_lstm, X_test_lstm, y_train, y_test, label_encoder

X_train_cnn, X_test_cnn, X_train_lstm, X_test_lstm, y_train, y_test, label_encoder = load_and_preprocess_data()

# üìå 3. **CNN Modelini Tanƒ±mla**
from tensorflow.keras.optimizers import Adam

def build_cnn_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(64, (5, 5), activation='relu', padding="same", input_shape=input_shape),
        BatchNormalization(),
        MaxPooling2D((2, 1)),
        Conv2D(128, (5, 5), activation='relu', padding="same"),
        BatchNormalization(),
        MaxPooling2D((2, 1)),
        Conv2D(256, (3, 3), activation='relu', padding="same"),
        BatchNormalization(),
        MaxPooling2D((2, 1)),
        Conv2D(512, (3, 3), activation='relu', padding="same"),  # üî• **Ekstra katman ekledik!**
        BatchNormalization(),
        MaxPooling2D((2, 1)),
        Flatten(),
        Dense(1024, activation='relu'),  # üî• **Daha b√ºy√ºk fully connected katman**
        Dropout(0.4),
        Dense(512, activation='relu'),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])

    optimizer = Adam(learning_rate=0.00001)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# üìå 4. **LSTM Modelini Tanƒ±mla**
def build_lstm_model(input_shape, num_classes):
    model = Sequential([
        Bidirectional(LSTM(512, return_sequences=True, activation='tanh', dropout=0.4), input_shape=input_shape),  # üî• **Daha b√ºy√ºk LSTM katmanƒ±**
        Bidirectional(LSTM(256, return_sequences=True, activation='tanh', dropout=0.3)),
        LSTM(128, activation='tanh', dropout=0.2),
        Dense(512, activation='relu'),  # üî• **Daha b√ºy√ºk fully connected katman**
        Dropout(0.3),
        Dense(128, activation='relu'),
        Dropout(0.2),
        Dense(num_classes, activation='softmax')
    ])

    optimizer = Adam(learning_rate=0.000001)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# üìå 5. **CNN Modelini Eƒüit ve Kaydet**
with tf.device('/GPU:0'):
    cnn_model = build_cnn_model(input_shape=(X_train_cnn.shape[1], 1, 1), num_classes=len(label_encoder.classes_))
    cnn_history = cnn_model.fit(X_train_cnn, y_train, epochs=100, batch_size=16, validation_data=(X_test_cnn, y_test))
    cnn_model.save("cnn_health_model.h5")
    print("\n‚úÖ CNN Modeli Kaydedildi: cnn_health_model.h5")

# üìå 6. **LSTM Modelini Eƒüit ve Kaydet**
with tf.device('/GPU:0'):
    lstm_model = build_lstm_model(input_shape=(X_train_lstm.shape[1], 1), num_classes=len(label_encoder.classes_))
    lstm_history = lstm_model.fit(X_train_lstm, y_train, epochs=100, batch_size=16, validation_data=(X_test_lstm, y_test))
    lstm_model.save("lstm_health_model.h5")
    print("\n‚úÖ LSTM Modeli Kaydedildi: lstm_health_model.h5")

# üìå 7. **Eƒüitim ve Test S√ºreleri**
cnn_test_time = time.time()
cnn_predictions = np.argmax(cnn_model.predict(X_test_cnn), axis=1)
cnn_test_time = time.time() - cnn_test_time

lstm_test_time = time.time()
lstm_predictions = np.argmax(lstm_model.predict(X_test_lstm), axis=1)
lstm_test_time = time.time() - lstm_test_time

print(f"\n‚è≥ CNN Test Time: {cnn_test_time:.4f} seconds")
print(f"‚è≥ LSTM Test Time: {lstm_test_time:.4f} seconds")

def plot_confusion_matrix(y_true, y_pred, title):
    conf_matrix = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12, 8))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Tahmin Edilen")
    plt.ylabel("Ger√ßek")
    plt.show()

# üìå 8. **Model Tahminleri ve Confusion Matrix √áizimi**
cnn_predictions = np.argmax(cnn_model.predict(X_test_cnn), axis=1)
lstm_predictions = np.argmax(lstm_model.predict(X_test_lstm), axis=1)

plot_confusion_matrix(y_test, cnn_predictions, "CNN - Damage Type")
plot_confusion_matrix(y_test, lstm_predictions, "LSTM - Damage Type")

# üìå 9. **Accuracy & Loss Grafikleri**
def plot_training_history(history, model_name):
    plt.figure(figsize=(12, 5))

    # Accuracy Grafiƒüi
    plt.subplot(1, 2, 1)
    plt.plot(history.history["accuracy"], label="Train Accuracy")
    plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
    plt.title(f"{model_name} Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()

    # Loss Grafiƒüi
    plt.subplot(1, 2, 2)
    plt.plot(history.history["loss"], label="Train Loss")
    plt.plot(history.history["val_loss"], label="Validation Loss")
    plt.title(f"{model_name} Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()

    plt.show()

plot_training_history(cnn_history, "CNN")
plot_training_history(lstm_history, "LSTM")
```

```python
import os
import numpy as np
import pandas as pd
import time
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, LSTM, Dropout, MaxPooling2D, BatchNormalization, Bidirectional
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# üìå 1. **GPU Kullanƒ±mƒ±nƒ± Etkinle≈ütir**
print("TensorFlow GPU Kullanƒ±labilirliƒüi:", tf.config.list_physical_devices('GPU'))
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print("‚úÖ GPU belleƒüi dinamik olarak ayarlandƒ±.")
    except RuntimeError as e:
        print("‚ùå GPU bellek ayarƒ± yapƒ±lamadƒ±:", e)

os.environ["CUDA_VISIBLE_DEVICES"] = "0"

# üìå 2. **Veriyi Y√ºkle ve ƒ∞≈üle (Saƒülƒ±klƒ± + Hasarlƒ±)**
def load_and_preprocess_damage_data(input_path="dron_all_audio_features.csv"):
    data = pd.read_csv(input_path)

    # üìå **Eksik "Damage Type" deƒüerlerini "None" olarak ayarla (Saƒülƒ±klƒ± dronlar i√ßin)**
    data["Damage Type"].fillna("None", inplace=True)

    # **X ve Y ayrƒ±mƒ± (Hasar t√ºr√º sƒ±nƒ±flandƒ±rmasƒ± i√ßin)**
    X = data.drop(columns=["File Name", "Health Status", "Damage Type"]).values  # √ñzellikler
    y = data["Damage Type"].values  # √áƒ±kƒ±≈ü (Hasar T√ºr√º)

    # üìå **Etiketleri Sayƒ±sal Hale Getir (Label Encoding)**
    label_encoder = LabelEncoder()
    y = label_encoder.fit_transform(y)

    # **Veriyi stratified olarak eƒüitim ve test setlerine ayƒ±r**
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

    # üìå **CNN i√ßin veriyi uygun hale getir**
    X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], 1, 1)
    X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], 1, 1)

    # üìå **LSTM i√ßin veriyi uygun hale getir**
    X_train_lstm = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
    X_test_lstm = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

    return X_train_cnn, X_test_cnn, X_train_lstm, X_test_lstm, y_train, y_test, label_encoder

X_train_cnn, X_test_cnn, X_train_lstm, X_test_lstm, y_train, y_test, label_encoder = load_and_preprocess_damage_data()

# üìå 3. **CNN Modelini Tanƒ±mla**
from tensorflow.keras.optimizers import Adam

def build_cnn_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(64, (5, 5), activation='relu', padding="same", input_shape=input_shape),
        BatchNormalization(),
        MaxPooling2D((2, 1)),
        Conv2D(128, (5, 5), activation='relu', padding="same"),
        BatchNormalization(),
        MaxPooling2D((2, 1)),
        Conv2D(256, (3, 3), activation='relu', padding="same"),
        BatchNormalization(),
        MaxPooling2D((2, 1)),
        Flatten(),
        Dense(512, activation='relu'),
        Dropout(0.3),
        Dense(256, activation='relu'),
        Dropout(0.2),
        Dense(num_classes, activation='softmax')
    ])

    optimizer = Adam(learning_rate=0.0001)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# üìå 4. **LSTM Modelini Tanƒ±mla**
def build_lstm_model(input_shape, num_classes):
    model = Sequential([
        Bidirectional(LSTM(256, return_sequences=True, activation='tanh', dropout=0.3), input_shape=input_shape),
        Bidirectional(LSTM(128, return_sequences=True, activation='tanh', dropout=0.3)),
        LSTM(64, activation='tanh', dropout=0.1),
        Dense(256, activation='relu'),
        Dropout(0.3),
        Dense(28, activation='relu'),
        Dropout(0.2),
        Dense(num_classes, activation='softmax')
    ])

    optimizer = Adam(learning_rate=0.00001)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# üìå 5. **CNN Modelini Eƒüit ve Kaydet**
with tf.device('/GPU:0'):
    cnn_model = build_cnn_model(input_shape=(X_train_cnn.shape[1], 1, 1), num_classes=len(label_encoder.classes_))
    cnn_history = cnn_model.fit(X_train_cnn, y_train, epochs=100, batch_size=32, validation_data=(X_test_cnn, y_test))
    cnn_model.save("cnn_damage_model_with_healthy.h5")
    print("\n‚úÖ CNN Modeli Kaydedildi: cnn_damage_model_with_healthy.h5")

# üìå 6. **LSTM Modelini Eƒüit ve Kaydet**
with tf.device('/GPU:0'):
    lstm_model = build_lstm_model(input_shape=(X_train_lstm.shape[1], 1), num_classes=len(label_encoder.classes_))
    lstm_history = lstm_model.fit(X_train_lstm, y_train, epochs=100, batch_size=32, validation_data=(X_test_lstm, y_test))
    lstm_model.save("lstm_damage_model_with_healthy.h5")
    print("\n‚úÖ LSTM Modeli Kaydedildi: lstm_damage_model_with_healthy.h5")

# üìå 7. **Confusion Matrix √áizimi**
def plot_confusion_matrix(y_true, y_pred, title):
    conf_matrix = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12, 8))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
                xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title(f"Confusion Matrix - {title}")
    plt.xlabel("Tahmin Edilen")
    plt.ylabel("Ger√ßek")
    plt.show()

# üìå 8. **Model Tahminleri ve Confusion Matrix √áizimi**
cnn_predictions = np.argmax(cnn_model.predict(X_test_cnn), axis=1)
lstm_predictions = np.argmax(lstm_model.predict(X_test_lstm), axis=1)

plot_confusion_matrix(y_test, cnn_predictions, "CNN - Damage Type (Healthy + Damaged)")
plot_confusion_matrix(y_test, lstm_predictions, "LSTM - Damage Type (Healthy + Damaged)")

# üìå 9. **Accuracy & Loss Grafikleri**
def plot_training_history(history, model_name):
    plt.figure(figsize=(12, 5))

    # Accuracy Grafiƒüi
    plt.subplot(1, 2, 1)
    plt.plot(history.history["accuracy"], label="Train Accuracy")
    plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
    plt.title(f"{model_name} Accuracy")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()

    # Loss Grafiƒüi
    plt.subplot(1, 2, 2)
    plt.plot(history.history["loss"], label="Train Loss")
    plt.plot(history.history["val_loss"], label="Validation Loss")
    plt.title(f"{model_name} Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()

    plt.show()

plot_training_history(cnn_history, "CNN")
plot_training_history(lstm_history, "LSTM")
```
